```python
# remove NAs, this should be ok for the initial model building, 
# but then we should recover incomplete entries. 
# For the testing, before remove NAs, we should drop the column SeriousDlqin2yrs,
# otherwise dropna would drop everything
df_noNA = df.dropna()
df_testing_red = df_testing.drop('SeriousDlqin2yrs',1)
df_testing_noNA = df_testing_red.dropna()
```

```python
#Parameters for the lasso model, different parameters can be use to setup the best model
# and then combine them all together
alpha = 0.1
lasso = Lasso(alpha=alpha,tol=0.000000001)
```

```python
# fit
is_deq = lasso.fit(df_noNA[['RevolvingUtilizationOfUnsecuredLines','age','NumberOfTime30-59DaysPastDueNotWorse',
                            'DebtRatio','MonthlyIncome','NumberOfOpenCreditLinesAndLoans','NumberOfTimes90DaysLate',
                            'NumberRealEstateLoansOrLines','NumberOfTime60-89DaysPastDueNotWorse','NumberOfDependents']]
                   ,df_noNA[['SeriousDlqin2yrs']])
```

```python
# prediction
prediction = is_deq.predict(df_testing_noNA[['RevolvingUtilizationOfUnsecuredLines','age','NumberOfTime30-59DaysPastDueNotWorse',
                            'DebtRatio','MonthlyIncome','NumberOfOpenCreditLinesAndLoans','NumberOfTimes90DaysLate',
                            'NumberRealEstateLoansOrLines','NumberOfTime60-89DaysPastDueNotWorse','NumberOfDependents']])
# add the prediction as a column in the testing noNA
df_testing_noNA['prediction']=prediction
```

```python
#Create csv file. If the row was dropped in the preparation stage, 
# 0.5 is assigned to the predicted value.
# to do: add the string "Id,Probability" as header of the csv

for i in df_testing.index:
    if i in df_testing_noNA['Unnamed: 0'] :
#        [prediction] = is_deq.predict(df_testing_noNA[['RevolvingUtilizationOfUnsecuredLines','age','NumberOfTime30-59DaysPastDueNotWorse',
#                            'DebtRatio','MonthlyIncome','NumberOfOpenCreditLinesAndLoans','NumberOfTimes90DaysLate',
#                            'NumberRealEstateLoansOrLines','NumberOfTime60-89DaysPastDueNotWorse','NumberOfDependents']].ix[i])
        my_prediction = df_testing_noNA['prediction'].ix[i]
        if my_prediction > 0 :
            myline = "%d,%7.8f\n" %(i+1,my_prediction)
        else :
            myline = "%d,%7.8f\n" %(i+1,0.0)
    else :
        myline = "%d,%7.8f\n" %(i+1,0.5)
    
    with open("predictions.csv","a") as myfile:
        myfile.write(myline)
```
